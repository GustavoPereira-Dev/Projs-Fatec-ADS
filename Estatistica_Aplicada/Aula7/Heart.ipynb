**8 - Qual a probabilidade de sortear ao acaso uma pessoa do sexo feminino com idade inferior a 35 anos?**

```python
sortear_menor_35 = len(heart[heart['Age'] < 35]) / len(heart) # 2.28% de sortear uma pessoa com idade inferior a 35
sortear_menor_35
```

```python
sortear_mulher = len(heart[heart['Sex'] == "F"]) / len(heart) # 21.02% a possibilidade de sortear uma mulher
sortear_mulher
```

```python
sortear_mulher - sortear_menor_35 # probabilidade de sortear uma pessoa do sexo feminino com idade inferior a 35 anos
```

```python
# prompt: Qual a probabilidade de sortear ao acaso uma pessoa do sexo feminino com idade inferior a 35 anos?
# Resposta correta
# Calculando a probabilidade de uma pessoa ser mulher E ter menos de 35 anos.
mulheres_menos_35 = len(heart[(heart['Sex'] == 'F') & (heart['Age'] < 35)])
probabilidade = mulheres_menos_35 / len(heart)
print(f"A probabilidade de sortear uma mulher com menos de 35 anos √©: {probabilidade:.4f}")
```
